```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,47,40,46,54]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(x,y,color="green", linewidth=5, linestyle="solid")
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(x,y,"+--")
```

```python
days = [1,2,3,4,5,6,7]
max_t = [43,45,67,56,67,89,73]
min_t = [56,55,67,46,77,59,63]
avg_t = [33,35,37,56,57,69,73]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("weather")

plt.plot(days, max_t, label="max", color = "blue")
plt.plot(days, min_t, label="min", color = "red")
plt.plot(days, avg_t, label="average")

plt.legend(loc="best")
```
